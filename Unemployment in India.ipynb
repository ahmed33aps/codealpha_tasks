COVID-19 Impact and Trends
Lockdown Spike: Immediately following the lockdown start in late March 2020, both sectors experienced a sharp, unprecedented surge in unemployment.

Peak Levels: Urban unemployment peaked significantly higher than rural, reaching nearly 30% in the months following the lockdown. Rural unemployment also rose sharply, peaking at over 20%.

Baseline Stability: Prior to the pandemic (mid-2019 to early 2020), the unemployment rate was relatively stable, with Urban rates consistently hovering around 10–12% and Rural rates between 7–10%.

Regional Volatility: The shaded areas (confidence intervals) indicate that while the average trend was high, certain regions experienced even more extreme fluctuations, as evidenced by your earlier data showing a maximum rate of 76.74%.

Area Comparison
Urban Vulnerability: The Urban sector was more severely affected by the lockdown than the Rural sector, likely due to the total halt of service and manufacturing industries.

Rural Resilience: While the Rural unemployment rate increased, it remained lower than the Urban rate throughout the crisis, potentially supported by continued agricultural activity.

Policy Insights
Targeted Urban Support: The data suggests that economic safety nets specifically for urban workers are crucial during nationwide lockdowns.

Recovery Monitoring: The sharp decline visible toward July 2020 indicates a rapid but partial recovery as initial restrictions were eased.
